## Task 2: 
### Some of the time series are stationary, some have a complex structure, some show a downward trend: Identify the **ridership types**  that have steadily increased in popularity,  and that have steadily decreased. by how much? (e.g what is the ratio of usage in the first 10 and last 10 weeks)

```python
mta_swipe = np.nansum(mta, axis=0)/600
mta_swipe = np.transpose(mta_swipe)
print (mta_swipe.shape)
mtadf_swipe = pd.DataFrame(mta_swipe)
mtadf_swipe.index = pd.Index(pd.date_range("2010-05-21", "2014-02-8", freq="W"))
ridetype = np.array(['10t', '14d', '1d', '2t', '30d', '7d', 'ada', 'afas', 'exp','ez',
            'ff','mc','mr','month','rfm','rr','sen','spec','tcmc','t',
            'umlim','xbus','stud'])
mtadf_swipe.columns = ridetype
mtadf_swipe.head(10)
```

```python
fig2 = plt.figure(figsize=(20,30))
fig2.suptitle('Trend for each fare type', fontsize=28, fontweight='bold')
x = 0
for i in ridetype:
    y = pd.rolling_mean(mtadf_swipe[i],50)
    ax = fig2.add_subplot(6,4,x+1)
    ax.plot(mtadf_swipe.index,y);
    ax.set_title(i)
    ax = fig2.autofmt_xdate()
    x = x+1
```

### Comment

Based on visual inspection, fare types that are under steady increases are: **14d, 7d, rr, sen, tcmc**. Fare types that are under decrease: **month**.
